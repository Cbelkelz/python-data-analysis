```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.facecolor"]="white"
plt.rcParams["figure.figsize"]=(7,7)

data = pd.read_excel(r"C:\Users\Christabel\Desktop\the nigeria data (cleaned).xlsx")
```

```python
data.set_index("S/N",inplace=True)
```

```python
data
```

```python
data.drop(data.iloc[:,1:57], axis=1, inplace=True)
data
```

```python
for x in data.loc[:, 2015:].columns:
    data[x] = pd.to_numeric(data[x], errors="coerce")
```

```python
data.dtypes
```

```python
data
```

```python
a=data.loc[(data["Indicator Name"]=="Population, female")|(data["Indicator Name"]=="Population, male")]
```

```python
a
```

```python
a.set_index("Indicator Name", inplace=True)
```

```python
a.T.plot(kind="bar", xlabel="Year", ylabel="Population(millions), male & female")
plt.title('Nigeria Population, male & female 2015-2020', fontsize=18, fontweight='bold',color='b' )
```

```python
b=data.loc[(data["Indicator Name"]=="Imports of goods and services (current LCU)")|(data["Indicator Name"]=="Exports of goods and services (current LCU)")]
b.set_index("Indicator Name", inplace=True)
b.T
```

```python
b.T.plot(kind="bar", xlabel="Year", ylabel="Export & Import of goods and sevices")
plt.title('Nigeria Import & Export from 2015-2020', fontsize=18, fontweight='bold',color='r' )
```

```python
c=data.loc[(data["Indicator Name"]=="Merchandise exports (current US$)")|(data["Indicator Name"]=="Merchandise imports (current US$)")]
c.set_index("Indicator Name", inplace=True)
c.T.plot(kind="pie", subplots=True, figsize=(15,15), labeldistance=1.2, autopct="%1.2f%%")
plt.title('Import and Export Mechhandise', color='k')
```

```python
d=data.loc[(data["Indicator Name"]=="Urban population")|(data["Indicator Name"]=="Rural population")]
d.set_index("Indicator Name", inplace=True)
d.T
```

```python
d.T.plot(kind="bar", color=["g","r"])
plt.xlabel("Year", fontweight="bold", fontsize=14)
plt.ylabel("Urban & Rural Population", fontweight="bold", fontsize=14)
plt.title('Nigeria Urban & Rural Population', fontsize=18, fontweight='bold',color='r' )
```
