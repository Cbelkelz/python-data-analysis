```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
a = pd.DataFrame({
    "Income": [1350000,2150000,1570000,1730000,2010000,3210000,5540000],
    "Expendicture": [950000,1850000,1350000,1400000,2900000,3650000,1730000],
    "Year": [2005, 2006, 2007, 2008, 2009,2010,2011]})
a
```

```python
a.set_index("Year", inplace=True)
a
```

```python
a["Surplus"] = a["Income"] - a["Expendicture"]
```

```python
a
```

```python
a.plot(kind="bar", color=["g","r","y"],figsize=(10,10))
plt.ylabel("Income,Expendicture,Surplus")
plt.title("Company xyz yearly cash flow")
plt.minorticks_on()
#plt.xticks(np.arange(0,7,0.1))
```

```python
Income =[1350000,2150000,1570000,1730000,2010000,3210000,5540000]
Expenditure =[950000,1850000,1350000,1400000,2900000,3650000,1730000]
Year =[2005, 2006, 2007, 2008, 2009,2010,2011]
Surplus=[]
for x in range(len(Income)):
    a = Income[x]-Expenditure[x]
    Surplus.append(a)
Surplus
```

```python
plt.figure(figsize=[10,10])
plt.plot(Year,Income,label="Income")
plt.plot(Year,Expenditure,label="Expenditure")
plt.plot(Year,Surplus,label="Surplus")
plt.legend()
```

```python
plt.figure(figsize=[10,10])
plt.plot(Year,Income, color="green",label="Income")
plt.plot(Year,Expenditure,color="r",label="Expenditure")
plt.plot(Year,Surplus,  color="k", marker="x", label="Surplus")

plt.suptitle("Company XYZ yearly cashflow", fontsize=18, fontweight="bold", fontfamily="Times New Roman", color="g")
plt.legend()
plt.minorticks_on()
plt.grid(which="both")
```

```python
plt.bar(Year,Income, color="green",label="Income")
plt.bar(Year,Expenditure,color="r",label="Expenditure",alpha=0.5)

plt.suptitle("Company XYZ yearly cashflow", fontsize=18, fontweight="bold", fontfamily="Times New Roman", color="g")
plt.legend()
plt.minorticks_on()
plt.grid(which="both")
```
