```python
import pandas as pd
files = r"C:\Users\Christabel\Desktop\Python materials\DA_materials\stocks101.xlsx"
x=pd.read_excel(files)
x
```

```python
x["S/N"]= [a for a in range(1,16)]
x.set_index("S/N", inplace = True)
x
```

```python
x.groupby("Month")["Value"].max()
```

```python
x.groupby("Month")["Value"].min()
```

```python
e =pd.Series(x.groupby("Month")["Value"].max(), name = "Maximum Value")
f =pd.Series(x.groupby("Month")["Value"].min(), name = "Minimum Value")
newef =pd.concat([e,f],axis = 1)
newef
```

```python
x.groupby("Stock")["Value"].max()
```

```python
x.groupby("Stock")["Value"].min()
```

```python
min_stock =pd.Series(x.groupby("Stock")["Value"].min(), name = "Minimum stock")
max_stock =pd.Series(x.groupby("Stock")["Value"].max(), name = "Maximum stock")
newline = pd.concat([min_stock,max_stock], axis = 1)
newline
```
